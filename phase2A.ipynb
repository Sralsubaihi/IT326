# pre-processeing data

removing outliers:

```python
import numpy as np
import pandas as pd

df = pd.read_csv("Stroke_Datas.csv")

num_colums = ["bmi", "avg_glucose_level","age"]

Q1 = df[num_colums].quantile(0.25)
Q3 = df[num_colums].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df[num_colums] < lower_bound) | (df[num_colums] > upper_bound)).sum()
print("Number of outliers per column (IQR Method):")
print(outliers)

print("------")
df_cleaned = df[~((df[num_colums] < lower_bound) | (df[num_colums] > upper_bound)).any(axis=1)]
print(f"Number of rows before removing outliers: {df.shape[0]}")
print(f"Number of rows after removing outliers: {df_cleaned.shape[0]}")


df_cleaned.to_csv("cleaned-data.csv", index=False)
```

used IQR to detected outliers - capping outliers to the nearest value , without renoving

```python
import numpy as np
import pandas as pd

df = pd.read_csv("Stroke_Datas.csv")


num_colums = ["bmi", "avg_glucose_level","age"]

Q1 = df[num_colums].quantile(0.25)
Q3 = df[num_colums].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df[num_colums] < lower_bound) | (df[num_colums] > upper_bound)).sum()
print("Number of outliers per column:")
print(outliers)

print("\n---capping data into the nearest value---")

for col in num_colums:
    df[col] = np.where(df[col] < lower_bound[col], lower_bound[col], df[col])
    df[col] = np.where(df[col] > upper_bound[col], upper_bound[col], df[col])


outliers = ((df[num_colums] < lower_bound) | (df[num_colums] > upper_bound)).sum()

print("\nNumber of outliers per column:")

print(outliers)

print("\n---after capping data---")

df_cleaned = df[~((df[num_colums] < lower_bound) | (df[num_colums] > upper_bound)).any(axis=1)]


df_cleaned.to_csv("cleaned_stroke_data.csv", index=False)

```

we used the (**IQR**) method to detect and handle outliers in numerical columns such as BMI, average glucose level, and age. Initially, **110 outliers** were identified in the BMI column and **627 outliers** in the average glucose level column, while no outliers were found in the age column. Instead of removing these outliers, which could lead to data loss, we applied a **capping technique** to adjust extreme values to the nearest acceptable range.

This means that values below the lower bound were replaced with the lower bound, and values above the upper bound were replaced with the upper bound. Capping ensures that the dataset remains intact while reducing the impact of extreme values, leading to better model stability and accuracy. After capping, a final check confirmed that all outliers were successfully handled, making the dataset cleaner and more suitable.

##feature selection:

```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold


df = pd.read_csv("cleaned_stroke_data.csv")

print("Original Dataset:")
print(df.head())

df_numeric = df.select_dtypes(include=["number"]).dropna()

selector = VarianceThreshold(threshold=0.2)
X_new = selector.fit_transform(df_numeric)

selected_features = [df_numeric.columns[i] for i in range(df_numeric.shape[1]) if selector.get_support()[i]]

print("\nSelected Features after Variance Threshold:")
print(df[selected_features].head())

```

We applied the **Variance Threshold method** to select the most relevant numerical features in the dataset. Features with very low variance contribute little to distinguishing between different classes, making them less useful. By setting a variance threshold of 0.2, we removed features with minimal variability, ensuring that only the most informative features were retained. This process helped to focus on meaningful data while eliminating redundant or constant features, leading to a cleaner, more efficient dataset.
